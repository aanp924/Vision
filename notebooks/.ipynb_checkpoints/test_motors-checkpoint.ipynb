#### Test Ultrasonic sensors

```python
vBot.Ctrl_Ultra_Sensor(1)#open
time.sleep(1) 
diss_H = vBot.read_data_array(0x1b,1)[0]
diss_L = vBot.read_data_array(0x1a,1)[0]
dis = diss_H<< 8 | diss_L 
print(str(dis) + "mm") 
vBot.Ctrl_Ultra_Sensor(0)#close
```
